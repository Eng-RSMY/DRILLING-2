```python
from tkinter import *
import math
```

```python
window=Tk()
window.geometry("550x200")
window.title("Well Trajectory Calculator")
l1=Label(window,text='North increment')
l1.grid(row=0,column=0)
l2=Label(window,text='BUR (deg/100ft)')
l2.grid(row=0,column=2)
l3=Label(window,text='Kick off point')
l3.grid(row=2,column=0)
l4=Label(window,text='Total Vertical Depth')
l4.grid(row=1,column=2)
l5=Label(window,text='East increment')
l5.grid(row=1,column=0)
l6=Label(window,text='Length of tangential')
l6.grid(row=2,column=2)
north_test=StringVar()
e1=Entry(window,textvariable=north_test)
e1.grid(row=0,column=1)

bur_test=StringVar()
e2=Entry(window,textvariable=bur_test)
e2.grid(row=0,column=3)

kop_test=StringVar()
e3=Entry(window,textvariable=kop_test)
e3.grid(row=2,column=1)

TVD_test=StringVar()
e4=Entry(window,textvariable=TVD_test)
e4.grid(row=1,column=3)

east_test=StringVar()
e5=Entry(window,textvariable=east_test)
e5.grid(row=1,column=1)

tang_test=StringVar()
e6=Entry(window,textvariable=tang_test)
e6.grid(row=2,column=3)

def azt():
    north=float(e1.get())
    east=float(e5.get())
    az=math.atan2(east,north)
    Azimuth=math.degrees(az)      
    f1=Label(text=Azimuth)
    f1.grid(row=4,column=1)        
       
    
def hzt():
    north=float(e1.get())
    east=float(e5.get())
    hzt=((north**2)+(east**2))**0.5
    f2=Label(text=hzt)
    f2.grid(row=5,column=1)
    
def r():
    bur=float(e2.get())
    radius= 5730/bur
    f3=Label(text=radius)
    f3.grid(row=6,column=1)

def inc():
    bur=float(e2.get())
    radius= 5730/bur
    north=float(e1.get())
    east=float(e5.get())
    hzt=((north**2)+(east**2))**0.5
    tvd=float(e4.get())
    kop=float(e3.get())
    x= radius-hzt
    y=tvd-kop
    z= (x**2 + y**2)**0.5
    k=radius/z
    j=math.asin(k)
    g=math.atan2(x,y)
    j1=math.degrees(j)
    g1=math.degrees(g)
    inclination=j1-g1
    f4=Label(text=inclination)
    f4.grid(row=4,column=3)
    
def L():
    bur=float(e2.get())
    radius= 5730/bur
    north=float(e1.get())
    east=float(e5.get())
    hzt=((north**2)+(east**2))**0.5
    tvd=float(e4.get())
    kop=float(e3.get())
    x= radius-hzt
    y=tvd-kop
    z= (x**2 + y**2)**0.5
    k=radius/z
    j=math.asin(k)
    g=math.atan2(x,y)
    j1=math.degrees(j)
    g1=math.degrees(g)
    inclination=j1-g1
    bur=float(e2.get())
    curve=inclination*100/bur
    f5=Label(text=curve)
    f5.grid(row=5,column=3)

def md():
    bur=float(e2.get())
    radius= 5730/bur
    north=float(e1.get())
    east=float(e5.get())
    hzt=((north**2)+(east**2))**0.5
    tvd=float(e4.get())
    kop=float(e3.get())
    x= radius-hzt
    y=tvd-kop
    z= (x**2 + y**2)**0.5
    k=radius/z
    j=math.asin(k)
    g=math.atan2(x,y)
    j1=math.degrees(j)
    g1=math.degrees(g)
    inclination=j1-g1
    bur=float(e2.get())
    curve=inclination*100/bur
    kop=float(e3.get())
    tangent=float(e6.get())
    md=kop+tangent+curve
    f6=Label(text=md)
    f6.grid(row=6,column=3)
    
    
b1=Button(window,text='Quadrant Azimuth (0-90 deg)',relief=RAISED,command=azt)
b1.grid(row=4,column=0)
b2=Button(window,text='Horizontal Departure',relief=RAISED,command=hzt)
b2.grid(row=5,column=0)
b3=Button(window,text='Radius of curvature',relief=RAISED,command=r)
b3.grid(row=6,column=0)
b4=Button(window,text='Final inclination',relief=RAISED,command=inc)
b4.grid(row=4,column=2)
b5=Button(window,text='Build up section length',relief=RAISED,command=L)
b5.grid(row=5,column=2)
b6=Button(window,text='Total Measured depth',relief=RAISED,command=md)
b6.grid(row=6,column=2)

window.mainloop()
```
